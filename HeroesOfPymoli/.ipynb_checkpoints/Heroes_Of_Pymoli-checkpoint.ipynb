```python
import pandas as pd
import numpy as np
import os

file ='/Users/marciooliver/Desktop/BootCamp/Boot/02-Homework/04-Pandas/Instructions/HeroesOfPymoli/Resources/purchase_data.csv'

file_df = pd.read_csv(file)
file_df.head()
```

```python
#Question 1
#Total Number of Players

# Unique screen players names and finding the length of that list

unique_players_count = len(file_df.drop_duplicates(['SN']))

# Print player count
total_players_df = pd.DataFrame([{'Total Players': unique_players_count}])

#Grooming - get rid of the index number 
total_players_df.set_index('Total Players', inplace = True)
total_players_df
```

```python
#Question 2
#Purchasing Analysis (Total)
#Get rid of duplicates
no_dupli_items = file_df.drop_duplicates(['Item ID'])

#Number of Unique Items
total_unique_items = len(no_dupli_items)

#Total Number of Purchases
total_num_pur = file_df['Price'].count()

#Total Revenue
total_rev = round(file_df['Price'].sum(),2)

#Average Purchase Price
avg_pur_price = round(total_rev/total_num_pur, 2)

pur_analysis = pd.DataFrame([{
    
    'Number of Unique Items': total_unique_items,
    'Average Purchase Price': avg_pur_price,
    'Total Purchases': total_num_pur,
    'Total Revenue': total_rev
}])

#Grooming -  formating values and get rid of the index number
format_mapping={'Total Revenue': '${:,.2f}', 'Average Purchase Price': '${:.2f}', 'Total Purchases': '${:,.2f}'}

for key, value in format_mapping.items():
    pur_analysis[key] = pur_analysis[key].apply(value.format)

pur_analysis.set_index('Number of Unique Items', inplace = True)
pur_analysis
```

```python
#Question 3
#Gender Demographics

#Get rid of duplicates
no_dupli_players = file_df.drop_duplicates(['SN'])

#gender_counts
gender_count = no_dupli_players.groupby(['Gender']).size().to_frame('Count').reset_index()

#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed
gender_count['Percentage of Players'] = (gender_count['Count']/unique_players_count) * 100


#Grooming -  formating values and get rid of the index number
gender_count['Percentage of Players'] = gender_count['Percentage of Players'].apply('{:.1f}%'.format)

gender_count.set_index('Gender', inplace=True)

gender_count
```

```python
#Question 4
#Purchasing Analysis (Gender) - The below each broken by gender


#Data with duplicates
gender_total = file_df.groupby(['Gender'])

# Creating Variables
#Average Purchase Price
#Total Purchase Value
pur_count = gender_total['SN'].count()

pur_price = gender_total['Price'].mean()

pur_value = gender_total['Price'].sum()

# Average Purchase Total per Person
pur_norm = (pur_value / gender_count['Count'])

# Merging variables, creating dict
pur_ana_gen = pd.DataFrame({'Purchase Count': pur_count,
                            'Average Purchase Price': pur_price,
                            'Average Purchase per Person': pur_norm,
                            'Total Purchase Value': pur_value
                           })

#Grooming -  formating values and get rid of the index number
format_mapping={'Average Purchase Price': '${:,.2f}',
                'Total Purchase Value': '${:,.2f}',
                'Average Purchase per Person': '${:,.2f}'}

for key, value in format_mapping.items():
    pur_ana_gen[key] = pur_ana_gen[key].apply(value.format)


pur_ana_gen


```

```python
#Question 5
#Age Demographics - The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

age_raw = file_df
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_raw['Age Group'] = pd.cut(age_raw['Age'], age_bins, labels=age_group)

age_broken = age_raw.groupby(['Age Group'])
per_count_age = age_broken.SN.nunique()
#Purchase Count
pur_count_age = age_broken['Price'].count()
#Total Purchase Value
pur_value_age = age_broken['Price'].sum()
#Average Purchase Price
ave_pur_value_age = age_broken['Price'].mean()
#Average Purchase Total per Person by Age Group'
ave_pur_per_age = (pur_value_age /per_count_age)


pur_ana_age = pd.DataFrame({'Purchase Count': pur_count_age,
                                'Average Purchase Price': ave_pur_value_age,
                                'Average Purchase per Person': ave_pur_per_age,
                                'Total Purchase Value': pur_value_age
                           })

#Grooming -  formating values and get rid of the index number
format_mapping={'Average Purchase Price': '${:,.2f}',
                'Total Purchase Value': '${:,.2f}',
                'Average Purchase per Person': '${:,.2f}'}

for key, value in format_mapping.items():
    pur_ana_age[key] = pur_ana_age[key].apply(value.format)


pur_ana_age
```

```python
# Question 6 - Top Spenders

# Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

# Purchase Count
# Average Purchase Price
# Total Purchase Value

spender_df = file_df.groupby('SN').Price.agg(count=('count'),
                                    average_value=('mean'),
                                    total_value=('sum')
                                    ).rename(columns={'count':'Purchase Count',
                                                        'total_value':'Total Purchase Value',
                                                        'average_value': 'Average Purchase Value'
                                                        })

format_mapping={'Total Purchase Value': '${:,.2f}', 'Average Purchase Value': '${:.2f}'}

for key, value in format_mapping.items():
    spender_df[key] = spender_df[key].apply(value.format)


spender_df.sort_values(by=['Total Purchase Value'], ascending=False).head()
```

```python
#Question 7 - Most Popular Items

# Identify the 5 most popular items by purchase count, then list (in a table):


# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

items_df = file_df.groupby('Item ID').agg(name=('Item Name','first'),
                                        count=('Price','count'),
                                        item_price=('Price', 'first'),
                                        total_value=('Price','sum')
                                        ).rename(columns={'count':'Purchase Count',
                                                        'name':'Item Name',
                                                        'item_price': 'Item Price',
                                                        'total_value':'Total Purchase Value'
                                                        })



format_mapping={'Total Purchase Value': '${:,.2f}', 'Item Price': '${:.2f}'}

for key, value in format_mapping.items():
    items_df[key] = items_df[key].apply(value.format)


items_df.sort_values(by=['Purchase Count'], ascending=False).head()
```

```python
# Question 8 - Most Profitable Items

# Identify the 5 most profitable items by total purchase value, then list (in a table):

# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

items_df = file_df.groupby('Item ID').agg(name=('Item Name','first'),
                                        count=('Price','count'),
                                        item_price=('Price', 'first'),
                                        total_value=('Price','sum')
                                        ).rename(columns={'count':'Purchase Count',
                                                        'name':'Item Name',
                                                        'item_price': 'Item Price',
                                                        'total_value':'Total Purchase Value'
                                                        })

items_df_sorted= items_df.sort_values(by=['Total Purchase Value'], ascending=False)

format_mapping={'Total Purchase Value': '${:,.2f}', 'Item Price': '${:.2f}'}

for key, value in format_mapping.items():
    items_df_sorted[key] = items_df_sorted[key].apply(value.format)


items_df_sorted.head()


```
